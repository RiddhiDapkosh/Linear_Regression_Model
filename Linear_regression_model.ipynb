```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Experience Years', y='Salary', data=df)
plt.title('Scatter Plot of Experience Years vs Salary')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
x=df[["Experience Years"]]
y=df["Salary"]
plt.figure(figsize=(10,4))
plt.scatter(x,y)
plt.grid()
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title("Scatter Plot of Experience Years vs Salary")
plt.show()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test ,y_train, y_test =train_test_split(x,y,test_size=0.7,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[10]])
```

```python
Year = float(input("Enter the number of experience years: "))
predicted_salary = lr.predict([[Year]])
print(predicted_salary)
```

```python
x=df[["Experience Years"]]
y=df["Salary"]
plt.figure(figsize=(10,4))
plt.scatter(x,y)
plt.grid()
plt.plot(x,lr.predict(x),color='red')
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title("Scatter Plot of Experience Years vs Salary")
plt.show()
```
